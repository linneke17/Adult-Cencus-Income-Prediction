**ADULT CENCUS INCOME**
====

**Overview of the "Adult Census Income" Dataset:**

The "Adult Census Income" dataset, sourced from Kaggle, provides a rich collection of demographic and socioeconomic information about individuals. Collected from the United States Census, this dataset encompasses a variety of features that allow for an in-depth analysis of income patterns within the population.


The dataset includes information such as age, education, marital status, occupation, gender, race, capital gains and losses, hours worked per week, and native country. Each entry in the dataset represents an individual, making it valuable for studying the factors that contribute to income disparities.


Key Features in the Dataset:

    Demographic Information: Age, gender, race, and native country provide insights into the diverse characteristics of the population.

    Educational Background: The dataset includes both categorical and numerical representations of an individual's education, allowing for a detailed examination of the impact of education on income.

    Employment Details: Workclass and occupation shed light on an individual's professional life, providing a basis for understanding income sources.

    Marital and Family Status: Marital status and relationship status provide information about an individual's family structure, which can influence financial stability.

    Financial Metrics: Capital gains, capital losses, and hours worked per week offer financial indicators that contribute to the overall income profile.

**Objectives of Analyzing the "Adult Census Income" Dataset:**

The primary objectives of analyzing this dataset are aligned with addressing income inequality and contributing to Sustainable Development Goals (SDGs):

    Gender-Based Income Analysis:
        Investigate and analyze income disparities between men and women within a specific geographical area.
        Explore the factors contributing to income differences, with a focus on gender-related variables such as occupation, education, and marital status.

    SDG 8 - Decent Work and Economic Growth:
        Contribute to SDG 8, specifically targeting Target 8.5, which aims to ensure full and productive employment and decent work for all, regardless of gender, age, or disability.
        Address Indicator 8.5.1 by monitoring average hourly earnings among different groups.

    SDG 10 - Reducing Inequalities:
        Align with SDG 10, particularly Target 10.2, which aims to empower and promote social, economic, and political inclusion for all, regardless of characteristics such as age, gender, and socioeconomic status.
        Address Indicator 10.2.1 by tracking the proportion of people living below 50 percent of the median income, stratified by sex, age groups, and disability status.

By exploring and interpreting the patterns within this dataset, the analysis aims to contribute valuable insights into income distribution, with a special emphasis on gender-related factors, fostering a more equitable understanding of economic participation and opportunities within the studied population.

Splitting the dataset into training and test sets before diving into exploratory data analysis (EDA) and feature engineering is crucial for building robust and reliable machine learning models.

The training set is analogous to a textbook, providing the model with foundational knowledge, while the test set serves as the exam to assess how well the model has learned. If you mix them up too soon, the model might essentially memorize the answers instead of understanding the underlying patterns.

By keeping the test set separate, it's like having a final exam that the model has never seen before, providing a more realistic assessment of its performance on new, unseen data. This precautionary measure ensures that the model doesn't inadvertently 'cheat' during its learning phase, resulting in a more accurate evaluation of its predictive capabilities when faced with unfamiliar information.

Handling Numeric Data (Transformation & Normalization)
===


Based on the conducted Exploratory Data Analysis (EDA), it is essential to normalize the numerical data in this dataset to make it suitable for inclusion in a model. We have chosen a feature-wise normalization approach, where the normalization is tailored to the characteristics of each column. Specifically, the 'age,' 'fnlwgt,' and 'education.num' columns have been normalized using a log transformation due to their uneven data distribution, as observed in the histograms during the EDA phase. On the other hand, we opted for RobustScaler to normalize data with outliers, aiming to mitigate the impact of outliers.

In simpler terms, after examining the data distribution through EDA, we found it necessary to adjust the scale of the numerical features for them to be suitable for modeling. We applied a feature-wise normalization method, considering the unique characteristics of each column. For columns like 'age,' 'fnlwgt,' and 'education.num,' we used a log transformation to handle their uneven data distribution. Meanwhile, for columns with outliers, we employed RobustScaler to normalize the data, reducing the influence of outliers in the process. This approach ensures that the data is appropriately prepared for further analysis or modeling.

```python
#ROBUST SCALER
from sklearn.preprocessing import RobustScaler

# Selecting numeric columns for robust scaling
numeric_columns = ['capital.gain', 'capital.loss', 'hours.per.week']

scaler = RobustScaler()

x_train[numeric_columns] = scaler.fit_transform(x_train[numeric_columns])
x_test[numeric_columns] = scaler.fit_transform(x_test[numeric_columns])
```

```python
# Log transformation
columns_for_log_transform = ['age', 'fnlwgt', 'education.num']

x_train[columns_for_log_transform] = np.log(x_train[columns_for_log_transform] + 1)
x_test[columns_for_log_transform] = np.log(x_test[columns_for_log_transform] + 1)
```

**Frequency Encoding**

Frequency Encoding for 'workclass', 'marital.status', 'occupation', 'relationship', 'race', 'native.country':


Frequency encoding is employed for these categorical columns to represent each category based on its frequency in the dataset. This encoding method assigns a numerical value to each category proportional to its occurrence. This approach is suitable when dealing with categorical features with a large number of unique values, as it captures the prevalence of each category in the dataset. By using frequency encoding, we transform the categorical values into meaningful numerical representations that reflect their distribution in the data.

In the default configuration of the BaggingClassifier, the base classifier is DecisionTreeClassifier. Multiple decision trees are trained on different subsets of the training data, and their predictions are combined through majority voting for classification tasks. This ensemble approach helps mitigate the variance of individual models, resulting in a more stable and reliable overall model.

**MODEL 8: STACKING**

Stacking is an ensemble learning technique that combines multiple base classifiers to improve overall predictive performance. In this specific case, the stacking classifier utilizes two different base classifiers: a **Random Forest classifier (rf)** and a **Support Vector Classifier (svc)**. The Random Forest classifier is an ensemble of decision trees, and the Support Vector Classifier is a linear model with the added benefit of feature scaling using a standard scaler. These diverse base classifiers capture different aspects of the data and may excel in different situations.

**The meta-model, in this case, is a Logistic Regression classifier (LogisticRegression()),** which acts as the final estimator to make the ultimate prediction based on the outputs of the base classifiers. The stacking classifier combines the predictions of the base classifiers and feeds them into the meta-model to generate the final prediction.

```python
# Stacking
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import StackingClassifier
import warnings
warnings.filterwarnings('ignore')


estimators = [
    ('rf', RandomForestClassifier(n_estimators = 10, random_state = 42)),
    ('svc', make_pipeline(StandardScaler(), LinearSVC(random_state=42)))
]

meta_model = LogisticRegression()

stacking_class = StackingClassifier(estimators = estimators, final_estimator = meta_model)
stacking_class.fit(x_train,y_train)
```

```python
y_predict = stacking_class.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print('\nClassification Report (Stacking)\n')
print(classification_report(y_test, y_predict, target_names = ['0', '1']))
```

The Stacking model, incorporating a Random Forest (RF) and a Support Vector Classifier (SVC), exhibits solid performance in classifying individuals based on their income status. For instances where income is less than or equal to 50k, the model achieves an impressive 87% precision, correctly identifying 94% of the actual instances and yielding an F1-score of 90%.

On the other hand, for individuals with income greater than 50k, the model maintains a respectable 75% precision, correctly identifying 54% of the actual instances, resulting in a 63% F1-score. The overall accuracy of 85% underscores the model's ability to make accurate predictions across both income classes. The macro average, considering an unweighted mean of precision, recall, and F1-score, stands at 77%, while the weighted average, considering the number of true instances for each class, is 84%. This suggests that the Stacking model provides a balanced trade-off between accuracy and generalization, making it a robust choice for predicting income status in the given dataset.
